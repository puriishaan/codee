```python
import serial
import json
import re 
```

```python


# Open the serial port (replace 'COM4' with your Arduino's port)
ser = serial.Serial('COM4', 9600)  # Adjust 'COM4' to match your port

m_data_list = []
t_data_list = []

# Regex pattern for incoming serial data
pattern = re.compile(r"T(?P<t>[\d.]+)x(?P<x>\d+)y(?P<y>\d+)z(?P<z>\d+)")

# Time interval for automatic saving (in seconds)
save_interval = 10  # Save data every 10 seconds
last_save_time = time.time()

def get_magnet_data():
    # Read and decode the incoming serial data
    line = ser.readline().decode('utf-8').strip()

    # Debug: print the raw data being read
    print(f"Raw data received: {line}")

    # Match the line with the regular expression
    match = pattern.match(line)

    if match:
        data = match.groupdict()
        data_type = 'T' if 't' in data else 'M'

        if data_type == 'M':
            m_data_list.append(data)
            print(f"Magnetometer Data Added: {data}")  # Debug: Confirmation that data was added
            return 'M', data
        elif data_type == 'T':
            t_data_list.append(data)
            print(f"Temperature Data Added: {data}")  # Debug: Confirmation that data was added
            return 'T', data
    else:
        print(f"Unexpected data format: {line}")
        return None

def save_data():
    # Save M and T data to separate JSON files
    with open('magnetometer_data.json', 'w') as m_file:
        json.dump(m_data_list, m_file, indent=4)
    with open('temperature_data.json', 'w') as t_file:
        json.dump(t_data_list, t_file, indent=4)
    print("Data automatically saved to magnetometer_data.json and temperature_data.json")

# Continuously collect data
while True:
    magnet_data = get_magnet_data()

    if magnet_data is not None:
        data_type, data = magnet_data

        # Store M and T data separately
        if data_type == "M":
            print(f"Magnetometer Data: {data}")
        elif data_type == "T":
            print(f"Temperature Data: {data}")

    # Check if it's time to save data
    current_time = time.time()
    if current_time - last_save_time >= save_interval:
        save_data()
        last_save_time = current_time  # Reset the timer after saving
```
